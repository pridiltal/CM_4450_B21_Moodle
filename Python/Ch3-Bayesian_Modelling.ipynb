# Chapter 3: Bayesian Modelling

In the previous chapter, we explored Bayesian computation, focusing on both Markovian methods (such as Markov Chain Monte Carlo) and non-Markovian methods (such as variational inference). These approaches provided powerful tools to approximate posterior distributions when exact solutions are not feasible.

However, these methods raise an important question:

**What kind of models require such sophisticated computational techniques?**

This chapter addresses that question by focusing on Bayesian modelling—the process of constructing probabilistic representations of real-world phenomena.

In Bayesian modelling, we begin by specifying:

- A prior distribution to represent existing knowledge,
- A likelihood function to describe the data-generating process,
- And observed data,

which together define the posterior distribution—the central object computed using the methods introduced earlier.

The distinction between Markovian and non-Markovian computation also reflects the complexity of the models we build. Simple models may allow direct or efficient sampling, while more complex models such as those with high-dimensional parameters or latent structures—require advanced computational strategies.

A key feature of Bayesian modelling is that uncertainty is explicitly modelled. Parameters are treated as random variables, and this allows us to move beyond point estimates toward full probabilistic inference.

In this chapter, we will develop a range of Bayesian models, starting from simple and interpretable structures and gradually moving toward more flexible and expressive frameworks. These include:

- Linear models, forming the foundation of many statistical methods,
- Mixture models, which capture hidden subpopulations,
- Latent variable models such as topic models,
- State space models, for sequential and time-dependent data,
- Gaussian processes, enabling non-parametric modelling,
- And smoothing with basis expansions, linking Bayesian ideas with functional approximation.

The emphasis will be on understanding how to formulate models, interpret their components, and connect them to real-world problems, while leveraging the computational tools developed in the previous chapter.

## Bayesian Linear Models

We begin our study of Bayesian modelling with one of the most fundamental and widely used model classes: linear models.

At its core, a linear model assumes a simple but powerful idea:

**The response variable can be expressed as a linear combination of input variables, plus some uncertainty.**

In a classical setting, linear regression focuses on estimating fixed coefficients that best explain the relationship between predictors and the response. However, this approach provides only a single “best” estimate and does not fully capture uncertainty in the model parameters.

The Bayesian perspective takes this a step further by treating these coefficients as random variables. Instead of estimating a single value, we assign:

- Prior distributions to the coefficients, representing our initial beliefs,
- A likelihood function based on the assumed data-generating process (often Gaussian noise),

and then combine them to obtain a posterior distribution over the coefficients.

This shift has important consequences:

- We obtain a **distribution of plausible parameter values**, not just point estimates,
- We can quantify uncertainty in predictions,
- And we can naturally incorporate prior knowledge into the model.

Bayesian linear models also serve as a foundation for many more advanced models. Concepts introduced here, such as prior specification, likelihood construction, and posterior inference, extend directly to more complex settings like hierarchical models, Gaussian processes, and neural networks.

Many problems in science, engineering, business, and artificial intelligence involve understanding relationships between variables.

In these problems, we typically define:

- A **dependent variable (target/output)**: the variable we want to predict (usually continuous)
- One or more **independent variables (inputs/features)**: variables used to explain or predict the target

The relationship between inputs and output can be written as:

$$y=f(x_1,x_2, \dots ,x_p) + \epsilon$$

where:

- $y$: dependent (target) variable
- $x_1,x_2, \dots ,x_p$: independent variables (inputs/features)
- $f(⋅)$: unknown function that represents the true relationship between inputs and output
- $\epsilon$: noise or random error (unexplained variation)

**What is $f$?**

The function $f$ represents the **true underlying relationship** between input variables and the output.

- It is generally unknown
- It is what we are trying to learn from data
- It may be linear or highly nonlinear in reality

In machine learning terms $f$ is the true data-generating process, which we attempt to approximate using a model.

Different modelling approaches correspond to different assumptions about $f$.

#### Linear Model Assumption

In linear regression, we assume that $f$ can be approximated by a linear function:

$$f(x_1,x_2, \dots ,x_p) = \omega_0 + \omega_1x_1 + \omega_2x_2 + \dots+\omega_px_p  $$


Therefore, the model becomes:

$$y = \omega_0 + \omega_1x_1 + \omega_2x_2 + \dots+\omega_px_p +\epsilon$$

where:

- $\omega_0$: intercept (bias term)
- $\omega_1 , \omega_2, \dots, \omega_p$: model parameters (weights)

### Simple linear regression

Now that we have discussed the general ideas behind linear regression, we now focus on how to formally build linear models.

A familiar starting point is the simple linear regression equation:

 $$y_i = \alpha+ \beta x_i +\epsilon_i$$

 This equation describes a linear relationship (straight-light relationship) between the input variable $x_i$ and the output variable $y_i$.

**Interpretation of Parameters**

- $\alpha$ (Intercept): The value of $y_i$ when $x_i=0$. Graphically, it is the point where the regression line intersects the y-axis.

- $\beta$ (Slope): Controls the rate of change of $y_i$ with respect to $x_i$. It represents the change in $y_i$ for a one-unit change in $x_i$

- $\epsilon_i$ (Error term): Represents random noise or unexplained variation in the data.

**Least Squares Estimation**

There are several methods to estimate the parameters α and β. One of the most common approaches is least squares fitting.

In practice, when fitting a regression line using software, this method is typically used internally.

The idea is to find parameter values that minimize the squared differences between observed and predicted values:

$$\min_{\alpha,\beta} \sum_{i=1}^{n} \left( y_i - \hat{y}_i \right)^2$$


Where 

$$\hat{y}_i = \hat{\alpha} + \hat{\beta} x_i$$

This formulation turns regression into an optimization problem, where we minimize a loss function.


**Probabilistic Interpretation**

Linear regression can also be expressed in a probabilistic framework.

Instead of modeling a deterministic line with noise, we assume that the output is generated from a probability distribution:

$$y_i \sim \mathcal{N}(\mu_i, \sigma^2)$$

where the mean is given by:

$$\mu_i=\alpha+\beta x_i$$

Therefore, the full model becomes:


$$y_i \sim \mathcal{N}(\alpha+\beta x_i, \sigma^2)$$

It is important to distinguish:

- $\epsilon_i$: individual random noise term (a random variable)
- $\sigma^2$: variance of the noise distribution

We Define 
$$\epsilon_i \sim \mathcal{N}(0, \sigma^2)$$

and therefore:

$$y_i=\alpha+\beta x_i+\epsilon_i$$

is equivalent to the probabilistic form above.

**Bayesian Formulation**

**Choice of Priors**

In a Bayesian framework, we treat the parameters as random variables and assign prior distributions:

$$\alpha \sim \mathcal{N}(\mu_{\alpha}, \sigma_{\alpha}^2)$$

$$\beta \sim \mathcal{N}(\mu_{\beta}, \sigma_{\beta}^2)$$

$$\sigma \sim \text{HalfNormal}(s)
\quad \text{or} \quad
\sigma \sim \text{HalfCauchy}(s)
$$



- Intercept ($\alpha$): Typically assigned a weakly informative (wide) Gaussian prior because its scale varies across problems.
- Slope ($\beta$): Also often given a weak Gaussian prior. In some cases, domain knowledge may inform its sign or range.
- Noise ($\sigma$): Must be positive, so distributions like Half-Normal or Half-Cauchy are commonly used. These act as regularizing priors.

If strong prior knowledge exists about variability, a Gamma distribution can also be used.

**Connection to Least Squares**

- When the priors are very weak (i.e., highly non-informative), the Bayesian solution converges closely to the least squares solution.

Thus:
- Least squares → point estimate
- Bayesian regression → full uncertainty-aware distribution
